```python
import numpy as np  ; import matplotlib.pyplot as plt   ; from scipy.optimize import curve_fit ; import seaborn as sns  ; XDATA = np.linspace(0, 4, 50)
def FUNC(x, a, b, c):       
    return  a *np.exp( -b*x) +c                     # _______________ curve_fit _______________
```

```python
YY = FUNC(XDATA, 2.5, 1.3, 0.5)                 ; y_noise = 0.2 * np.random.default_rng().normal(size=XDATA.size)            ; YDATA = YY + y_noise ; EE = 1  # 3
popt1, pcov1 = curve_fit(FUNC, XDATA, YDATA)    ; popt2, pcov2 = curve_fit(FUNC, XDATA, YDATA, bounds=(0, [EE, 1., 0.5]))    ; print(popt1) 
```
