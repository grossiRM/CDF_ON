<font color='black'> **Heat _____________________** </font> <br> <br> 
<font color='black'> 
$ 
\, - k\nabla^2T - S = 0
\,\,\,\,\,\,\,      \,\,\,\,\,\,\,
$
<font color='blue'> 
$ 
\, \mathbf{J} = - k\nabla^2T 
$

<font color='black'> **Discretization  | J |** </font>
<font color='black'> 
<br> <br>
<font color='blue'> 
$ 
\int_S \mathbf{J}\cdot\mathbf{n} dS    \approx \sum_{i=0}^{N_{ip}-1} \mathbf{J}_{ip}\cdot\mathbf{n}_{ip}A_{ip}
$
<font color='green'> 
$ 
\,\      = - \sum_{i=0}^{N_{ip}-1} k_{ip} \nabla T_{ip} \cdot\mathbf{n}_{ip}A_{ip}
$
<font color='orange'> 
$ 
\,\      = k \left. \frac{\partial T}{\partial x} \right|_w A_w - k \left. \frac{\partial T}{\partial x} \right|_e A_e                
$
<font color='blue'> 
$ 
\,\      = k \frac{T_P - T_W}{\Delta x_{WP}} A_w - k \frac{T_E - T_P}{\Delta x_{PE}} A_e
$

<font color='black'> **Discretized source term** </font>  <br>
<br>
<font color='blue'> 
$ 
\int_V S dV \approx S_P V_P       
\,\,\,\,\,\,\,      \,\,\,\,\,\,\,
$
<font color='green'> 
$ 
\,\,\,\,\,\,\,      r_P = - k \frac{T_E - T_P}{\Delta x_{PE}} A_e + k \frac{T_P - T_W}{\Delta x_{WP}} A_w - S_P V_P                
\,\,\,\,\,\,\,      \,\,\,\,\,\,\,
$
<font color='orange'> 
$ 
\,\,\,\,\,\,\,      r_P = F^d_e - F^d_w - S_P V_P
$

<font color='black'> **Source Convection and Radiation** </font>  <br>
<br>
<font color='blue'> 
$ 
\frac{S_{conv,P}}{V_P} = - h A_o (T_P - T_{\infty, c})
$
<font color='green'> 
$ 
\,\,\,\,\,\,\,      \,\,\,\,\,\,\,      \,\,\,\,\,\,\,      
\frac{S_{rad}}{V_P} = - \epsilon\sigma A_o (T_P^4 - T_{\infty,r}^4)
$

<font color='black'> **Diffusive fluxes through each face, $F^d$** </font>  <br>
<br>
<font color='blue'> 
$ 
F^d_e = - k \frac{T_E - T_P}{\Delta x_{PE}} A_e = - D_e \left(T_E - T_P\right)       
$
<font color='blue'> 
$ 
\,\,\,\,\,\,\,
\,\,\,\,\,\,\,      D_e = \frac{k A_e}{\Delta x_{PE}}
$
<font color='green'> 
$ 
\,\,\,\,\,\,\,\,\,\,\,\,\,\,    \,\,\,\,\,\,\,\,\,\,\,\,\,\,
\,\,\,\,\,\,\,      F^d_w = - k \frac{T_P - T_W}{\Delta x_{WP}} A_w = - D_w \left(T_P - T_W\right)                
$
<font color='green'> 
$ 
\,\,\,\,\,\,\,
\,\,\,\,\,\,\,      D_w = \frac{k A_w}{\Delta x_{WP}}
$

<font color='black'> **Cell residual ___**

\,\,\,\,\,\,\,
\,\,\,\,\,\,\,      r_P = D_w \left(T_P - T_W\right)
$
<font color='blue'>

\,\,\,\,\,\,\,  \,\,\,\,\,\,\,  \,\,\,\,\,\,\,
